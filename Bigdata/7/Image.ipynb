```python
from PIL import Image
import matplotlib.pyplot as plt
from numpy import asarray
from sklearn import cluster
import numpy as np

im = Image.open("rainbow.jpg")
display(im)
```

```python
data = asarray(im)
imshape = data.shape
plt.imshow(data)
```

```python
X = data.reshape(-1,3)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:,0],X[:,1],X[:,2],c= X/255.0, marker='.')
plt.show()
```

```python
# K-means
k = 10
k_mean_model = cluster.KMeans(n_clusters=k, max_iter=50, random_state=1)
k_mean_model.fit(X)
cluster_id = k_mean_model.labels_
centroids = k_mean_model.cluster_centers_
print(centroids)
```

```python
Xnew =  X.copy()
for i in range(k):
    Xnew[cluster_id==i,:] = centroids[i,:]
Xnew
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(X[:,0],X[:,1],X[:,2], c=Xnew/255.0, marker='.')
plt.show()
```

```python
Xnew = Xnew.reshape(imshape)
plt.imshow(Xnew)
```
