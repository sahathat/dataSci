```python
# imputation by change value from na to global mean, and 'default' size in outlet size
df_copy = df.copy()
df_copy
```

```python
# imputation
mean_weight = df_copy["Item_Weight"].mean()
df_copy["Item_Weight"] = df_copy["Item_Weight"].fillna(mean_weight)
df_copy["Outlet_Size"] = df_copy["Outlet_Size"].fillna("Default")
df_copy
```

# 3

```python
# use plotly express to plot graph
import plotly.express as px
colX= "Outlet_Size"
colY= "Item_Weight"
fig = px.histogram(df_copy, x=colX, y=colY,
             color="Item_Type",
             title="item weight by outlet size",
             labels={'Item_Weight':'weight','Outlet_Size':'size'},
             barmode='group'
            )
fig.show()
```
