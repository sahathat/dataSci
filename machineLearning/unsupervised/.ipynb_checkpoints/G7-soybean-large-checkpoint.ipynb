```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a pandas DataFrame called 'df'
# Replace this with your actual DataFrame
df = selected_df.copy()

# Calculate the count of missing values for each column
missing_count = df.isnull().sum()

# Calculate the total number of rows in the DataFrame
total_rows = len(df)

# Calculate the percentage of missing values for each column
missing_percentage = (missing_count / total_rows) * 100

# Create a DataFrame to hold missing values count and percentages
missing_data = pd.DataFrame({'Column': df.columns, 'MissingCount': missing_count, 'MissingPercentage': missing_percentage})

# Sort the DataFrame by MissingPercentage in descending order
missing_data = missing_data.sort_values('MissingPercentage', ascending=False)

# Create a figure with a single subplot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the count of missing data as a grouped bar chart
sns.barplot(x='Column', y='MissingCount', data=missing_data, ax=ax, palette='viridis', alpha=0.7)
ax.set_ylabel('Missing Count')
ax.set_title('Missing Data Summary')

# Add the percentage of missing data as a line chart on a secondary y-axis
ax2 = ax.twinx()
sns.lineplot(x='Column', y='MissingPercentage', data=missing_data, marker='o', color='orange', ax=ax2)
ax2.set_ylabel('Missing Percentage (%)', color='orange')

# Rotate x-axis labels to 45 degrees and align to the right
plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Encoder label

# diaporthe-stem-canker, charcoal-rot, rhizoctonia-root-rot,
# phytophthora-rot, brown-stem-rot, powdery-mildew,
# downy-mildew, brown-spot, bacterial-blight,
# bacterial-pustule, purple-seed-stain, anthracnose,
# phyllosticta-leaf-spot, alternarialeaf-spot,
# frog-eye-leaf-spot, diaporthe-pod-&-stem-blight,
# cyst-nematode, 2-4-d-injury, herbicide-injury.

i = 0
class_mapping = dict()
for cls in selected_df['class'].unique():
    class_mapping[cls] = i
    i = i + 1
selected_df['encode_class'] = selected_df['class'].map(class_mapping)
del selected_df['class']
print(selected_df['encode_class'])

# Handle missing values using kNN imputation
imputer = KNNImputer(n_neighbors=1)
feature_imputed = imputer.fit_transform(selected_df)
print(feature_imputed)

normalize = StandardScaler()
features = normalize.fit_transform(feature_imputed)
print(features)
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
from numpy import arange

# Create a linkage object.
methods = ['single','complete','average','ward']
linkage_col = dict()
for method in methods:
    # Create the figure and set it's size.
    plt.figure(figsize=(25,10))
    
    linkage_col[method] = linkage(features, method=method)

    # Make a dendrogram from the linkage object.
    dendrogram(linkage_col[method], orientation='top', distance_sort='descending')
    
    plt.xlabel("cluster data of soybean with 15 columns")
    plt.ylabel("distance")
    plt.title(f"soybean data with 15 columns with {method} linkage hierarchy clustering")
    plt.show()

    # Show the plot
    plt.show()
```

```python
# Determine the clusters using fcluster
max_d = [4,8,6,25]  # Adjust this value based on the dendrogram
i = 0
for method in methods:
    clusters = fcluster(linkage_col[method], t=max_d[i], criterion='distance')

    # Add the cluster labels to the DataFrame
    selected_df['Cluster'] = clusters

    # Plot the clusters
    plt.figure(figsize=(10, 6))

    # Create scatter plot for each cluster
    for cluster_id in selected_df['Cluster'].unique():
        cluster_data = features[selected_df['Cluster'] == cluster_id]
        plt.scatter(cluster_data[:,1], cluster_data[:,2], label=f'Cluster {cluster_id}')

    plt.xlabel("plant-stand")
    plt.ylabel("temp")
    plt.title(f"Clusters of Soybean in {method} mode")
    plt.legend()
    plt.show()
    i = i + 1
```
